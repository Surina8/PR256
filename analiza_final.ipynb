### Cenovna segmentacija (nizka • srednja • visoka)

**Kaj počnemo?**  
Izdelke razdelimo v tri cenovne razrede glede na tercile (tretjine) celotne porazdelitve cen:

| Segment | Pojasnilo |
|---------|-----------|
| **Nizka cena** | Izdelki v spodnji tretjini cenovne lestvice. |
| **Srednja cena** | Izdelki v srednji tretjini. |
| **Visoka cena** | Izdelki v zgornji tretjini. |

---

#### Box-plot vizualizacija

- **Škatla** prikazuje srednjih 50 % cen v posameznem segmentu (interkvartilni razpon, 25.–75. percentil).  
- **Vodoravna črta** v škatli označuje **mediano** cen.  
- **»Brki«** segajo do najnižjih in najvišjih cen, ki še niso izstopajoče.  
- **Pike** (če so) predstavljajo **izstopajoče** vrednosti.

Ker so segmenti že definirani *po ceni*, bo med njimi vedno jasna ločnica; box-ploti zato največ povedo o **razponu in razpršenosti** cen znotraj vsake skupine.

---

#### Ključna sporočila

1. **Hiter pregled** – takoj vidite, v kateri cenovni razred spada posamezen izdelek.  
2. **Primerjava razpršenosti** – lahko ocenite, kateri segment ima bolj strnjene (ali razpršene) cene.  
3. **Omejitve grafike** – graf ne kaže povezav s časom, trgovino ali vrsto izdelka; namenjen je zgolj notranji primerjavi cen.

```python
import pandas as pd
from rapidfuzz import process, fuzz

# 1. Load your data
df = pd.read_csv('vse_cene_grouped.csv', sep=';')
gov = pd.read_csv('SLO-cene.csv')

# Clean product names
gov['izdelek_clean'] = gov['izdelek'].str.strip().str.lower()
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Year'] = df['Date'].dt.year

years = [2018, 2019, 2020, 2021, 2022, 2023, 2024]

all_comparisons = {}

for year in years:
    print(f"\n=== Primerjava za leto {year} ===")
    # 1. Prepare government prices for the year
    gov_prices = gov[['izdelek_clean', str(year)]].rename(columns={str(year): 'SLO Price (€)'})
    gov_prices['SLO Price (€)'] = pd.to_numeric(gov_prices['SLO Price (€)'], errors='coerce')  # <-- ADD THIS LINE
    # 2. Get your average price per grouped product for the year
    df_year = df[df['Year'] == year]
    your_avg = df_year.groupby('Product Name Grouped')['Price (€)'].mean().reset_index()
    your_avg['Product Name Grouped Clean'] = your_avg['Product Name Grouped'].str.strip().str.lower()
    # 3. Fuzzy match
    matches = []
    for idx, row in gov_prices.iterrows():
        gov_name = row['izdelek_clean']
        match, score, _ = process.extractOne(
            gov_name,
            your_avg['Product Name Grouped Clean'],
            scorer=fuzz.token_sort_ratio
        )
        if score >= 60 and not pd.isna(row['SLO Price (€)']):
            your_price = your_avg[your_avg['Product Name Grouped Clean'] == match]['Price (€)'].values[0]
            matches.append({
                'SLO izdelek': row['izdelek_clean'],
                'SLO Price (€)': row['SLO Price (€)'],
                'Our Product': match,
                'Our Avg Price (€)': your_price,
                'Difference (€)': your_price - row['SLO Price (€)'],
                'Match Score': score
            })
    # 4. Create DataFrame for this year
    comparison = pd.DataFrame(matches)
    comparison = comparison.sort_values('Difference (€)', key=abs, ascending=False)
    all_comparisons[year] = comparison
    # 5. Show results for this year
    print(f"Primerjava cen za leto {year}:")
    display(comparison[['SLO izdelek', 'SLO Price (€)', 'Our Product', 'Our Avg Price (€)', 'Difference (€)', 'Match Score']].head(20))
```

```python
import matplotlib.pyplot as plt

# Povprečna razlika po letih (naš minus SLO)
avg_diffs = []
num_matches = []
for year in years:
    comp = all_comparisons[year]
    if not comp.empty:
        avg_diffs.append(comp['Difference (€)'].mean())
        num_matches.append(len(comp))
    else:
        avg_diffs.append(None)
        num_matches.append(0)

# 1. Povprečna razlika po letih (naš minus SLO)
plt.figure(figsize=(8,4))
plt.plot(years, avg_diffs, marker='o', color='royalblue')
plt.title('Povprečna razlika med našimi in SLO cenami po letih')
plt.xlabel('Leto')
plt.ylabel('Povprečna razlika (€)')
plt.axhline(0, color='gray', linestyle='--')
plt.grid(True)
plt.tight_layout()
plt.show()


# 4. Povzetek: koliko % naših cen je višjih/nižjih od SLO po letih
for year in years:
    comp = all_comparisons[year]
    if not comp.empty:
        pct_higher = (comp['Difference (€)'] > 0).mean() * 100
        pct_lower = (comp['Difference (€)'] < 0).mean() * 100
        print(f"Leto {year}: {pct_higher:.1f}% naših cen je višjih od SLO, {pct_lower:.1f}% nižjih.")
    else:
        print(f"Leto {year}: Ni ujemanj.")
```
